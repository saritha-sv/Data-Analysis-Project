<font size = 6px><b>Step: 5 Data Visualization </b></font>

```python
import matplotlib.pyplot as plt
```

```python
data.columns
```

<font size = 5px><b>Bar Graph</b></font>

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('job')['balance'].mean()
grouped.plot(kind='bar')
plt.title('Mean Balance by Job')
plt.xlabel('Job')
plt.ylabel('Mean Balance')
plt.show()
```

<font size = 5px><b>Horzontal Bar Graph</b></font>

```python
# Assuming 'data' is your DataFrame
grouped = data.head(20).groupby('marital')['balance']
grouped.plot(kind='barh')
plt.title('Distribution of Balance by Marital Status')
plt.ylabel('Balance')
plt.xlabel('Marital Status')
plt.show()
```

<h5><b>Pie Chart</b></h5>

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('education').size()

# Plotting the pie chart
grouped.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Education Levels')
plt.ylabel('')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

<h3><b>Histogram Graph</b></h3>

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('education')['balance']
grouped.plot(kind='hist', alpha=0.5, bins=20)
plt.title('Distribution of Balance by Education')
plt.xlabel('Balance')
plt.ylabel('Frequency')
plt.legend(title='Education')
plt.show()
```

<font size = 5px><b>line chart</b></font>

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('month')['duration'].mean()
grouped.plot(kind='line', marker='o')
plt.title('Mean Duration by Month')
plt.xlabel('Month')
plt.ylabel('Mean Duration')
plt.xticks(rotation=45)
plt.show()
```

<font size = 5px><b>Scatter Graph</b></font>

```python

# Assuming 'data' is your DataFrame
grouped = data.groupby('campaign')
for name, group in grouped:
    plt.scatter(group['duration'], group['balance'], label=name, alpha=0.5)
plt.title('Duration vs Balance by Campaign')
plt.xlabel('Duration')
plt.ylabel('Balance')
plt.legend(title='Campaign')
plt.show()
```

<font size = 6px><b>Step 6: Aggregate Functions</b></font>

```python
# groupby function for Customer ID and Quantity
df = pd.DataFrame(data)
grouped=df.groupby('job')['contact'].count()
print(grouped)
```

```python
# Assuming data is your DataFrame
aggregate_functions = {
    'balance': ['sum', 'mean', 'median', 'min', 'max', 'std', 'count'],    
}
summary_statistics = data.agg(aggregate_functions)
print(summary_statistics)
```
